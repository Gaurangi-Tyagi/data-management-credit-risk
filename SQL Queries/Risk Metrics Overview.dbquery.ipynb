```sql
SELECT 
    risk_segment,
    family_status,
    education_type,
    total_customers,
    defaulted_customers,
    default_rate_percent,
    ROUND(avg_credit_amount, 2) as avg_credit_amount,
    ROUND(avg_income, 2) as avg_income
FROM dm2_credit_risk.gold_credit_risk_metrics
ORDER BY default_rate_percent DESC;
```
