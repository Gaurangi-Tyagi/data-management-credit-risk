```sql
SELECT 
    family_status,
    SUM(total_customers) as total_customers,
    SUM(defaulted_customers) as defaulted,
    ROUND(SUM(defaulted_customers) / SUM(total_customers) * 100, 2) as default_rate
FROM dm2_credit_risk.gold_credit_risk_metrics
GROUP BY family_status
ORDER BY default_rate DESC
LIMIT 5;
```
